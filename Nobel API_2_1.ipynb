## Check Wikidata matches

```python
# pip install sparqlwrapper
# https://rdflib.github.io/sparqlwrapper/

import sys
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint_url = "https://query.wikidata.org/sparql"

query = """SELECT ?WDitem ?WDitemLabel ?WDNobelAPI ?WDdeath WHERE {
  ?WDitem wdt:P8024 ?WDNobelAPI.
  OPTIONAL {?WDitem wdt:P570 ?WDdeath}
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en". }
} order by xsd:integer(?WDNobelAPI)"""


def get_sparql_dataframe(endpoint_url, query):
    """
    Helper function to convert SPARQL results into a Pandas data frame.
    """
    user_agent = "salgo60/%s.%s" % (sys.version_info[0], sys.version_info[1])
 
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    result = sparql.query()

    processed_results = json.load(result.response)
    cols = processed_results['head']['vars']
    out = []
    for row in processed_results['results']['bindings']:
        item = []
        for c in cols:
            item.append(row.get(c, {}).get('value'))
        out.append(item)

    return pd.DataFrame(out, columns=cols)

WDNobel = get_sparql_dataframe(endpoint_url, query)
```

```python
WDNobel["WDNobelAPI"] = WDNobel["WDNobelAPI"].astype("int")
WDNobel.info()
```

```python
dfmerge = pd.merge(WDNobel, dftotPrize,how='outer', left_on="WDNobelAPI",right_on="id",indicator=True)
dfmerge['_merge'] = dfmerge['_merge'].str.replace('left_only','WD_only').str.replace('right_only','Nobel_only')
dfmerge.rename(columns={"_merge": "WD_Nobel"},inplace = True)    
#dfmerge[-10:]  
```
