```python
import urllib3, json
import pandas as pd  
from tqdm.notebook import trange  
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
urlbase = "https://kulturnav.org/api/search/entityType:Entity?start=" 

dftot = pd.DataFrame()
for i in trange(1,500000,20):
    url = urlbase + str(i)
    r = http.request('GET', url)
    try:
        data = json.loads(r.data)
        dftot = dftot.append(pd.DataFrame(data),sort=False)
    except:
        print("\tError in i=",i) 
```
